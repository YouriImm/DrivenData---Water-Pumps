```python
import numpy as np 
np.set_printoptions(suppress=True)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scikitplot as skplt
```

```python
#Read data
train_vals = pd.read_csv('../data/training_vals.csv')
train_labels = pd.read_csv('../data/training_labels.csv')
test_vals = pd.read_csv('../data/test_vals.csv')

#To make my train_test split easier later on: 
train_vals.set_index('id', inplace=True)
train_labels.set_index('id', inplace=True)

#For convenience
train_vals = train_vals.merge(train_labels, left_index=True, right_index=True)

train_vals.head()
```

```python
def target_label_visualization(dataset):
    fig, ax = plt.subplots(figsize=(8,6))
    ax.bar(x=dataset.status_group.value_counts().index, height=dataset.status_group.value_counts())
    ax.set_title("Target label distribution")
    ax.yaxis.grid(c='lightgrey')
    ax.set_axisbelow(True)
    ax.set_ylabel("Amount of waterpoints")
    rects = ax.patches

    for rect in rects: 
        x = rect.get_x() + (rect.get_width() / 2)
        y = rect.get_height()
        ax.text(x=x, y=y + 300, s=y, ha='center')
    plt.close()
    return fig
```

```python
target_label_visualization(train_vals)
```
